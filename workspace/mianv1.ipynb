```python
from ELTE import readRDF, ionForce
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
R = [1,1.5,2.5,3,4,5,6,7]
```

```python
f1 , na1 = ionForce("forces/force_ions_0_1_0.xyz",1)
f15 ,na15 = ionForce("forces/force_ions_0_15_0.xyz",1)
f25 ,na25 = ionForce("forces/force_ions_0_25_0.xyz",1)
f3 , na3 = ionForce("forces/force_ions_0_3_0.xyz",1)
f4 , na4 = ionForce("forces/force_ions_0_4_0.xyz",1)
f5 , na5 = ionForce("forces/force_ions_0_5_0.xyz",1)
f6 , na6 = ionForce("forces/force_ions_0_6_0.xyz",1)
f7 , na7 = ionForce("forces/force_ions_0_7_0.xyz",1)
```

```python
F_fluor = [f1,f15,f25,f3,f4,f5, f6, f7 ]
F_na = [na1,na15,na25,na3 ,na4 ,na5 ,na6 ,na7 ]
```

```python
plt.figure(figsize=(20,8))

plt.subplot(121)
plt.title("Force in between Fluorid atoms with at different positions")
plt.plot(R,F_fluor, "-o", label ="Fluorid")
plt.ylabel("F(r) ["+r'$\frac{kcal}{mol \cdot A}$'+"]")
plt.xlabel("r[A]")
plt.legend()

plt.subplot(122)
plt.title("Force in between Natrium atoms with at different positions")
plt.plot(R,F_na,"-o",label ="Natrium")
plt.ylabel("F(r) ["+r'$\frac{kcal}{mol \cdot A}$'+"]")
plt.xlabel("r[A]")
plt.legend()
plt.savefig("fig1.png")
```
